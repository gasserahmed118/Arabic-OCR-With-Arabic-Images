# 2. Imports


```python
import os
import cv2
import pytesseract
import easyocr
import matplotlib.pyplot as plt
```

# 3. Initialize Easy-OCR Reader

```python
reader = easyocr.Reader(['ar'])   
```

# 4. Reading Images

```python
image_folder = r"C:\Users\gasse\OneDrive\Desktop\10 OCR\OCR\Test_aracic_ocr"   

# Get all PNG files
image_paths = [os.path.join(image_folder, f) 
               for f in os.listdir(image_folder) if f.lower().endswith(".png")]

print("Found Images:")
for p in image_paths:
    print(" -", p)

print("\n")
```

# 5. Loop Trrough Each Image

```python
for idx, img_path in enumerate(image_paths):

    # Load and convert
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Print shape
    print(f"IMAGE {idx+1}: {os.path.basename(img_path)}")
    print("Shape:", img_rgb.shape)
```

# 6. Run Pytesseract


# 6.1 Image 1

```python
img = cv2.imread(image_paths[0])
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

text = pytesseract.image_to_string(img_rgb, lang="ara")

print("Extracted Text from Image 1:")
print(text.strip())
print("-" * 60)

plt.imshow(img_rgb)
plt.title("Image 1")
plt.axis("off")
```

# 7. EASY-OCR


# 7.1 Image 1

```python
print("--- EASY OCR RESULT: IMAGE 1 ---")

img = cv2.imread(image_paths[0])
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

easy_text = reader.readtext(img_rgb)

for (bbox, text, prob) in easy_text:
    print(text)

print("-" * 60)

plt.imshow(img_rgb)
plt.title("Image 1")
plt.axis("off")
```
